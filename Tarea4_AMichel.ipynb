```python
def dist_binomial(n,p):
    p = np.zeros(n)
    p[0]= (1 - p) ** n
    
    def llenar_vector(i)
        nonlocal p
        p[i + 1] = (((n - i)*p) / ((1 - p)*(i + 1 ))) * p[i]
        
    [llenar_vector(i) for i in range(n-1)]
    return p.cumsum()
```
